```python
# 통계 메서드

from random import *


# 난수 생성
data_1 = []
data_2 = []
for i in range(0, 10):
    data_1.append(uniform(0,1))
for i in range(0, 10):
    data_2.append(uniform(0,1))
    
# DataFrame 생성
dict_data = {
    'first' : data_1,
    'second' : data_2,
}

index_data = ['zero', '1st', '2st', '3st', '4st', '5st', '6st', '7st',  '8st', '9st']

df = pd.DataFrame(dict_data, index = index_data)
df   
```

```python
# 평균
df.mean()
df['first'].mean()
df['second'].mean()
```

```python
# 최댓값
df.max()
df['first'].max()
df['second'].max()
```

```python
# 최솟값
df.min()
df['first'].min()
df['second'].min()
```
